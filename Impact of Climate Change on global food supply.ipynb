```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_global_temp = pd.read_csv("D:\\Kaggle Project\\Analyzing the impact of climate change on global food supply\\annual_csv.csv")
```

```python
df_global_temp.head(10)
```

 Now lets filter our data so that we only get records corresponding to the NASA GISTEMP source

```python
df_global_temp = df_global_temp[df_global_temp['Source'] == 'GISTEMP'].reset_index()[["Source", "Year", "Mean"]]
```

```python
df_global_temp.head(10)
```

Next lets plot the mean annual temperature vs. time and making a line plot of the time series

```python
sns.set()
sns.lineplot(df_global_temp['Year'], df_global_temp['Mean'])
plt.ylabel("Mean")
plt.title("Average Global Mean Temperature")
```

```python
df_wheat = pd.read_csv("D:\\Kaggle Project\\Analyzing the impact of climate change on global food supply\\wheat-yields.csv")
df_wheat.head()
```

There are a total of 229 regions in rice production dataset and 163 regions in wheat production dataset. Developing regions are more vulnerable to the risks that climate change pose, it would be useful to narrow our scope. According to a article published in the Time Magazine in 2019, it stated that Nigeria, Haiti, Yemen, Philippines and Fiji will face the most severe consequences of climate change.

##### Let us first look into the rice and wheat production of Nigeria.

```python
df_rice_Nigeria = df_rice[df_rice['Entity']=='Nigeria'].reset_index()[["Entity", "Code", "Year", "Rice yield-hg/ha"]]
print(df_rice_Nigeria.head())
```

Next we can plot the crop yield for rice in Nigeria from 19602018

```python
sns.lineplot(df_rice_Nigeria['Year'], df_rice_Nigeria['Rice yield-hg/ha'])
plt.ylabel("Rice Production in Nigeria (hg/ha)")
plt.title("Annual Rice production in Nigeria")
```

Next we can overlay the global annual mean temperature and the crop yields for rice in Nigeria

```python
sns.set()
sns.lineplot(df_global_temp['Year'], df_global_temp['Mean'])
plt.ylabel("Mean")
plt.title("Average Global Mean Temperature and rice production in Nigeria")

sns.lineplot(df_rice_Nigeria['Year'], df_rice_Nigeria['Rice yield-hg/ha'])
plt.ylabel("Mean temperature/hg/ha rice in Nigeria ")
```

```python
df_wheat_Nigeria = df_wheat[df_wheat['Entity']=='Nigeria'].reset_index()[["Entity", "Code", "Year", "Wheat yield-hg/ha"]]
print(df_wheat_Nigeria.head())
```

```python
sns.lineplot(df_wheat_Nigeria['Year'], df_wheat_Nigeria['Wheat yield-hg/ha'])
plt.ylabel("Wheat Production in Nigeria (hg/ha)")
plt.title("Annual Wheat production in Nigeria")
```

```python
sns.set()
sns.lineplot(df_global_temp['Year'], df_global_temp['Mean'])
plt.ylabel("Mean")
plt.title("Average Global Mean Temperature and wheat production in Nigeria")

sns.lineplot(df_wheat_Nigeria['Year'], df_wheat_Nigeria['Wheat yield-hg/ha'])
plt.ylabel("Mean temperature/hg/ha wheat in Nigeria ")
```

There seems to be a significant drop in rice production in Nigeria between 1987 and 2006. The wheat production of Nigeria seem to hhave a downward trend with soignificant drop in production in 1990 and 2000.
